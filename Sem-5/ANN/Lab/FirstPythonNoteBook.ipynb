A backslash ( \ ) to split the logic across multiple lines. When the code is executed, Python will ignore the

#List creation and Slicing Notation: [start:stop:step]
start: The index to start from (defaults to the beginning of the list if not specified).
stop: The index to end at (defaults to the end of the list if not specified).
step: The step to take when slicing (a positive step moves forward, a negative step moves backward).

# Performing Data Preparation
The first step in the development of any deep learning model – after gathering the
data, of course – should be preparation of the data. The term Exploratory data analysis (EDA) means to
understand the data available, as well as to detect potential issues that may affect the
development of the model. steps involved

1. Import the required libraries.
2. Using pandas, load the .csv file.
3. Verify whether any qualitative data is present in the dataset.
4. Check for missing values. You can also add an additional sum() function to get the sum of missing values
in the entire dataset, without discriminating by column.
5. Check for outliers.
6. Separate the features from the target data.
7. Rescale the data using the standardization methodology.

# The .iloc[]
The .iloc[] method is used to select data from a DataFrame based 
on integer-based indexing (i.e., position-based selection rather than label-based).
1) The first argument is for selecting rows.
2) The second argument is for selecting columns.

```python
# Sample DataFrame for illustration
data = {
    'Appliances': [60, 70, 80, 90, 100],
    'lights': [10, 20, 30, 40, 50],
    'T1': [21.6, 22.1, 23.0, 24.5, 25.3],
    'RH_1': [40.5, 41.0, 42.0, 43.5, 44.3],
    'T2': [20.0, 21.2, 22.5, 23.6, 24.1]
}

df_data = pd.DataFrame(data)

# Describe using column names
describe_by_column = df_data[['Appliances', 'T1']].describe()

# Describe using column indices (0: 'Appliances', 1: 'lights')
describe_by_index = df_data.iloc[:, [0, 3]].describe()

# Print the results
print(f"Using column names:\n{describe_by_column}\n")
print(f"Using column indices:\n{describe_by_index}\n")
```
