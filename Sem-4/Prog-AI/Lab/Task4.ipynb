# Problem 01


```python
employee_records = []
def insert_employee_record(id, name, department, years_of_service):
    new_record = (id, name, department, years_of_service)
    employee_records.append(new_record)
def delete_employee_record(id):
    for record in employee_records:
        if record[0] == id:
            employee_records.remove(record)
            return
    print("Employee with ID {} not found.".format(id))
def update_years_of_service(id, new_years_of_service):
    for record in employee_records:
        if record[0] == id:
            record = list(record)
            record[3] = new_years_of_service
            employee_records[employee_records.index(record)] = tuple(record)
            return
    print("Employee with ID {} not found.".format(id))
def print_employee_records():
    print(f"ID \t Name \t Department \t Years of Service")
    for record in employee_records:
        print(f"{record[0]} \t {record[1]} \t {record[2]} \t {record[3]}")
while True:
    print("\nMenu:")
    print("1. Insert Employee Record")
    print("2. Delete Employee Record")
    print("3. Update Years of Service")
    print("4. Print Employee Records")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        id = input("Enter ID: ")
        name = input("Enter Name: ")
        department = input("Enter Department: ")
        years_of_service = input("Enter Years of Service: ")
        insert_employee_record(id, name, department, years_of_service)
    elif choice == '2':
        id = input("Enter ID of employee to delete: ")
        delete_employee_record(id)
    elif choice == '3':
        id = input("Enter ID of employee to update: ")
        new_years_of_service = input("Enter new Years of Service: ")
        update_years_of_service(id, new_years_of_service)
    elif choice == '4':
        print_employee_records()
    elif choice == '5':
        break
    else:
        print("Invalid choice. Please try again.")
```

# Problem 02

```python
reviews = []
def preprocess_review(review):
    review = review.lower()  
    review = ''.join(char for char in review if char.isalpha() or char.isspace())
    review = ''.join(char for char in review if not char.isdigit())
    return review
def insert_review(review):
    reviews.append(review)
def delete_review(review):
    if review in reviews:
        reviews.remove(review)
    else:
        print("Review not found in the collection.")
def update_review(old_review, new_review):
    if old_review in reviews:
        index = reviews.index(old_review)
        reviews[index] = new_review
    else:
        print("Review not found in the collection.")
def print_reviews():
    for index, review in enumerate(reviews, start=1):
        print(f"Review {index}: {review}")
while True:
    print("\nMenu:")
    print("1. Preprocess Review")
    print("2. Insert Review")
    print("3. Delete Review")
    print("4. Update Review")
    print("5. Print Reviews")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        review = input("Enter the review to preprocess: ")
        preprocessed_review = preprocess_review(review)
        print("Preprocessed Review:", preprocessed_review)
    elif choice == '2':
        review = input("Enter the review to insert: ")
        insert_review(review)
    elif choice == '3':
        review = input("Enter the review to delete: ")
        delete_review(review)
    elif choice == '4':
        old_review = input("Enter the review to update: ")
        new_review = input("Enter the new version of the review: ")
        update_review(old_review, new_review)
    elif choice == '5':
        print_reviews()
    elif choice == '6':
        break
    else:
        print("Invalid choice. Please try again.")
```
